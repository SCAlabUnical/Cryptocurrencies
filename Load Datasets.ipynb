Load Datasets:

```python
import pandas as pd
import os

cartella_sorgente = 'dfpyspark/dfpyspark1'
elenco_file = [f for f in os.listdir(cartella_sorgente) if f.endswith('.csv')]
chunk_caricati = []
for nome_file in elenco_file:
    percorso_file = os.path.join(cartella_sorgente, nome_file)
    chunk_df = pd.read_csv(percorso_file)
    chunk_caricati.append(chunk_df)

df_completo = pd.concat(chunk_caricati, ignore_index=True)
df_completo
```
